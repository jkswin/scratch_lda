```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# source https://www.youtube.com/watch?v=7LB1VHp4tLE 

samples = {"x":[1], "y":[-1]}

num_samples = 1000

for _ in range(num_samples):
    curr_y = samples["y"][-1]
    new_x = np.random.normal(curr_y/2, np.sqrt(3/4)) # compute a new value with one variable fixed
    new_y = np.random.normal(new_x/2, np.sqrt(3/4)) # rinse repeat
    samples["x"].append(new_x)
    samples["y"].append(new_y)

plt.scatter(**samples, c="orange", alpha=0.5, marker="o")
```
